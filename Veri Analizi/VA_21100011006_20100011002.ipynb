<h1>Aykiri veri analizi</h1>

```python
#boxplot cizdiren fonksiyon

def boxplot_column (df,column_name):
 sns.boxplot(data=df, x=column_name)
 plt.title(column_name)
# plt.savefig(column_name+'.png')
 plt.show()
```

```python
for column_name in numerical_columns:
 boxplot_column(data,column_name)
```

```python
#Aykırı değelerine en yakın aykırı olmayan değerler atmak


def manage_outlier(df,column_name):
 q3 = df[column_name].quantile(0.75)
 q1 = df[column_name].quantile(0.25)
 iqr = q3-q1
 lower_limit = q1-1.5*iqr
 upper_limit = q3+1.5*iqr

 df.loc[df[column_name]<lower_limit ,column_name] = lower_limit
 df.loc[df[column_name]>upper_limit ,column_name] = upper_limit

```

```python
for column_name in numerical_columns:
 manage_outlier(data,column_name)
```

```python
for column_name in numerical_columns:
 boxplot_column(data,column_name)
```

```python
# Benzer klinik profillere sahip erkek ve kadın hastalar arasında hayatta kalma oranları için lod-rank testi

from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test

# Benzer klinik profillere sahip hastaları seçme (örneğin, yaş, ejeksiyon fraksiyonu, serum kreatinin, serum sodyum, yüksek tansiyon ve anemi ile filtreleme)
criteria = (
    (data['age'] >= 50) & (data['age'] <= 70) &
    (data['ejection_fraction'] >= 30) & (data['ejection_fraction'] <= 50) &
    (data['serum_creatinine'] <= 2.0) &
    (data['serum_sodium'] >= 135) & (data['serum_sodium'] <= 145) &
    (data['high_blood_pressure'] == 1) &
    (data['anaemia'] == 1)
)

filtered_data = data[criteria]

# Erkek ve kadın hastalar olarak gruplandırma
male_data = filtered_data[filtered_data['sex'] == 1]
female_data = filtered_data[filtered_data['sex'] == 0]

# Kaplan-Meier hayatta kalma analizi
kmf_male = KaplanMeierFitter()
kmf_female = KaplanMeierFitter()

plt.figure(figsize=(10, 6))

kmf_male.fit(durations=male_data['time'], event_observed=male_data['DEATH_EVENT'], label='Erkek')
kmf_female.fit(durations=female_data['time'], event_observed=female_data['DEATH_EVENT'], label='Kadın')

ax = kmf_male.plot(ci_show=True)
kmf_female.plot(ax=ax, ci_show=True)

plt.title('Kaplan-Meier Hayatta Kalma Eğrileri: Erkek vs. Kadın (Benzer Klinik Profiller)')
plt.xlabel('Zaman (gün)')
plt.ylabel('Hayatta Kalma Olasılığı')
plt.legend()
plt.show()

# Log-rank testi
results = logrank_test(male_data['time'], female_data['time'], event_observed_A=male_data['DEATH_EVENT'], event_observed_B=female_data['DEATH_EVENT'])
print('Log-rank Testi p-değeri:', results.p_value)

if results.p_value < 0.05:
    print("Hayatta kalma oranlarında erkek ve kadın hastalar arasında istatistiksel olarak anlamlı bir fark vardır.")
else:
    print("Hayatta kalma oranlarında erkek ve kadın hastalar arasında istatistiksel olarak anlamlı bir fark yoktur.")
```

<h1>Değerlendirme ve Model kıyaslama</h1>

Bu aşamada Mcnemar testi kullanıldı
Neden Mcnemar testi?
 McNemar testi, ikili sınıflandırma problemlerinde model performanslarını karşılaştırmak için tercih edilir çünkü:

Eşleşmiş İkili Gözlemler: McNemar testi, aynı veri seti üzerinde iki farklı sınıflandırıcıyı test ederken eşleşmiş ikili gözlemler kullanır. Bu, her iki modelin de aynı veri noktalarını doğru veya yanlış sınıflandırdığı durumları dikkate alır, bu da sonuçları daha karşılaştırılabilir kılar.

Hata Korelasyonu: İki modelin aynı veri setindeki performansını karşılaştırırken, hata korelasyonunu göz önünde bulundurur. Bu, bağımsız örnekler varsayımının ihlal edilmediği anlamına gelir ve sonuçların daha güvenilir olmasını sağlar.

Basit ve Etkili: McNemar testi, kolay uygulanabilir ve yorumlanabilir bir istatistiksel testtir. Test, iki model arasındaki farkın anlamlı olup olmadığını hızlıca belirlemeyi sağlar.

Örneklem Büyüklüğü Gereksinimi: Diğer karmaşık testlere kıyasla, McNemar testi genellikle daha az veri gerektirir (bu örnekte 299), bu da küçük veri setlerinde bile uygulanabilirliğini artırır.

```python
import numpy as np
from statsmodels.stats.contingency_tables import mcnemar
from sklearn.metrics import accuracy_score

# Her model için tahminleri elde etmek
y_pred_best_clfdt = best_clfdt.predict(X_test)  # DecisionTree
y_pred_best_clfnb = best_clfnb.predict(X_test)  # NaiveBayes
y_pred_best_clfxg = best_clfxg.predict(X_test)  # XGboost
y_pred_best_clfex = best_clfex.predict(X_test)  # ExtraTreeClassifier

# Function to create the contingency table Mcnemar testi tablosu oluşturmak iki algoritma aynı doğru değer için
# İkisi doğru bulduğu duruşlar yani sum[model1_pred & model2_pred]

def get_contingency_table(y_true, y_pred1, y_pred2):
    table = np.zeros((2, 2), int)
    for true, pred1, pred2 in zip(y_true, y_pred1, y_pred2):
        if pred1 == true and pred2 == true:
            table[0, 0] += 1
        elif pred1 == true and pred2 != true:
            table[0, 1] += 1
        elif pred1 != true and pred2 == true:
            table[1, 0] += 1
        else:
            table[1, 1] += 1
    return table

# Algoritmalri ikişer şekilde kıyaslama
model_pairs = [
    (y_pred_best_clfdt, y_pred_best_clfnb, "DecisionTree vs NaiveBayes"),
    (y_pred_best_clfdt, y_pred_best_clfxg, "DecisionTree vs XGboost"),
    (y_pred_best_clfdt, y_pred_best_clfex, "DecisionTree vs ExtraTreeClassifier"),
    (y_pred_best_clfnb, y_pred_best_clfxg, "NaiveBayes vs XGboost"),
    (y_pred_best_clfnb, y_pred_best_clfex, "NaiveBayes vs ExtraTreeClassifier"),
    (y_pred_best_clfxg, y_pred_best_clfex, "XGboost vs ExtraTreeClassifier")
]


alpha = 0.05

#Her iki kiyaslama için sonuçlari
for y_pred1, y_pred2, description in model_pairs:
    contingency_table = get_contingency_table(y_test, y_pred1, y_pred2)
    result = mcnemar(contingency_table, exact=False, correction=True)
    if result.pvalue < alpha:
        print(f'{description}: p-value={result.pvalue:.4f}, Statistic={result.statistic} Different proportions of errors (reject H0)')
    else:
        print(f'{description}: p-value={result.pvalue:.4f}, Statistic={result.statistic} Same proportions of errors (fail to reject H0)')


models = {
    "best_clfdt": y_pred_best_clfdt,
    "best_clfnb": y_pred_best_clfnb,
    "best_clfxg": y_pred_best_clfxg,
    "best_clfex": y_pred_best_clfex
}

for model_name, y_pred in models.items():
    accuracy = accuracy_score(y_test, y_pred)
    print(f'{model_name} Accuracy: {accuracy:.4f}')
```
